### 3. Use a Subset of the data frame to create 'Price Label' Column using 'def' function

```python
# Create a Subset 
df = ords_prods_merge[:1000000]
```

```python
# Define the function 
def price_label(row):
    if row ['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range product'
    else: return 'Not enough data'
```

```python
# Apply the function 
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
# Check the count in the price range column 
df['price_range'].value_counts(dropna = False)
```

### 4. Using the loc function  to create 'Pricr range loc' column'

```python
# Using Loc for Low range products
ords_prods_merge.loc[ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low-range products'
```

```python
# Using Loc for mid range products
ords_prods_merge.loc[(ords_prods_merge['prices'] > 5) & (ords_prods_merge['prices'] <=15), 'price_range_loc'] = 'Mid-range Products'
```

```python
# Using Loc for high range products
ords_prods_merge.loc[ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range Products'
```

### 5. Using if statement with a loop to show the busiest day, least busy and regularly busy 

```python
# Using the if statement with a loop to show the busiest day, least busy and regularly busy 
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least Busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Creating a 'Busiest day' column in the data frame using result 
ords_prods_merge['Busiest_day'] = result
```

```python
# Count of the busy days in the dataframe 
ords_prods_merge['Busiest_day'].value_counts(dropna = False)
```

### 3. Create a label called "busiest_period_of_day

```python
# Identifying the busiest hours of the day
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

The criteria i will be using for the hour with the Most orders will be orders >2,000,000 which are 10,11,14,15,13,12,16,9,17. The criteria i will be using for fewest orders is anything <500,000 which aree 23,6,0,1,5,2,4,3. Average orders is everything in between which are 8,18,19,20,7,12,22

```python
# Create a column with the labels "Most Orders", "Average orders" and "Fewest orders"

result_3  = []
for value in ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9, 17]:
    result_3.append("Most Orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_3.append("Fewest Orders")
  else:
    result_3.append("Average Orders")
```

```python
# Adding 'period of the day' column into thhe dataframe 
ords_prods_merge['busiest_period_of_the_day'] = result_3
```
