# 3 Create loyalty flag for existing customers using transform and loc() function

```python
# Create a loyalty flag using the transform function to get the max number 
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_identity_number'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# If max order is > 40 that the loyalty flag shows loyal customer
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# If max order is euals to or less than 40 or greater than 10 that the loyalty flag shows regular customer 
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40)  & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
```

```python
# If max order is less than or equal to 10 that the loyalty flag shows new customer
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

```python
# Statics of the loyalty flag 
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.head()
```

# 4 Check for spending habit of each loyalty category 

```python
# Finding the statistics of the various types of customers under the loyalty flag 
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean', 'max', 'min']})
```

# 5 Targeting spenders in the marketing campaigns 

```python
# Create a loyalty flag using the transform function to get the mean number of prices 
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_identity_number'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(15)
```

```python
#If the average price of spending is less than 10 then its a low spender 
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'Spending' ] = 'Low Spender'
```

```python
# If the average price of spending is greater or  equals to 10 then its a high spender 
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'Spending' ] = 'High Spender'
```

```python
#Checking the frequency of the flag
ords_prods_merge['Spending'].value_counts(dropna = False)
```

# 7. Create a flag using the transform function to get the median 


```python
# Creating a flag for the median days since prior order for customers 
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_identity_number'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(15)
```

```python
# if the median days since prior order is more than 20 is a non frequent customer
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'order_regularity'] = 'Non-frequent customer'
```

```python
# if the median days since prior order is higher than 10 or lower than equals to 20 is a regular customer
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] > 10) & (ords_prods_merge['median_days_since_prior_order'] <= 20) , 'order_regularity'] = 'Regular customer'
```

```python
# if the median days since prior order is lower than or equals to 10 then its a frequent customer
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'order_regularity'] = 'Frequent customer'
```

```python
# Frequency of order regularity of customers
ords_prods_merge['order_regularity'].value_counts(dropna = False)
```
